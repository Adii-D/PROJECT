```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming 'Customer type' is a categorical variable
frequency_distribution_customer_type = df['Customer type'].value_counts()

# Define a dark color palette with unique patterns
colors = ['#1F77B4', '#FF7F0E', '#2CA02C', '#D62728']
patterns = ["//", "o", ".", "\\"]
hatches = [patterns[i % len(patterns)] for i in range(len(frequency_distribution_customer_type))]

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(9,5))

# Create bars with custom color palette and add patterns
bars = sns.barplot(x=frequency_distribution_customer_type.index, y=frequency_distribution_customer_type.values, palette=colors, hatch=hatches)

# Add data labels with custom format
for bar in bars.patches:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.01*max(frequency_distribution_customer_type.values), round(yval, 2), ha='center', va='bottom', color='white', fontsize=10)

# Customize the plot
plt.xlabel('Customer Type', fontsize=14, fontweight='bold', labelpad=15, color='black')
plt.ylabel('Frequency', fontsize=14, fontweight='bold', labelpad=15, color='black')
plt.title('Frequency Distribution of Customer Type', fontsize=16, fontweight='bold', pad=20, color='black')
plt.xticks(rotation=45, ha='right', color='black')
plt.yticks(fontsize=10, color='black')
sns.despine()

# Add a grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust space between plot and title
plt.subplots_adjust(top=0.85)

# Remove spines
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming 'Customer type' is a categorical variable
frequency_distribution_customer_type = df['Customer type'].value_counts()

# Define unique color palettes for each subplot
colors_list = [ ['#1F77B4', '#2CA02C'],
                ['#FF5733', '#FFC300'],
                ['#9933FF', '#00FF00'] ]

# Define unique patterns for each subplot
patterns_list = [ ["//", "o"],
                  ["x", "O"],
                  ["\\\\", "+"] ]

# Define unique labels for each subplot
labels_list = [ ['Type A', 'Type B'],
                ['Group X', 'Group Y'],
                ['Category 1', 'Category 2'] ]

# Create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(12, 6))

for i, (colors, patterns, labels) in enumerate(zip(colors_list, patterns_list, labels_list)):
    # Create bars with custom color palette and add patterns
    bars = sns.barplot(x=frequency_distribution_customer_type.index, y=frequency_distribution_customer_type.values, palette=colors, ax=axes[i], hatch=patterns)

    # Add data labels with custom format
    for bar, label in zip(bars.patches, labels):
        yval = bar.get_height()
        axes[i].text(bar.get_x() + bar.get_width()/2, yval + 0.01*max(frequency_distribution_customer_type.values), label, ha='center', va='bottom', color='white', fontsize=10, fontweight='bold')

    # Customize the plot
    axes[i].set_xlabel('Customer Type', fontsize=14, fontweight='bold', labelpad=15, color='black')
    axes[i].set_ylabel('Frequency', fontsize=14, fontweight='bold', labelpad=15, color='black')
    axes[i].set_title(f'Subplot {i+1}', fontsize=16, fontweight='bold', pad=20, color='black')
    axes[i].set_xticklabels(frequency_distribution_customer_type.index, rotation=45, ha='right', color='black')
    axes[i].tick_params(axis='y', labelsize=10, color='black')
    sns.despine(ax=axes[i])

    # Add a grid
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

    # Remove spines
    axes[i].spines['top'].set_visible(False)
    axes[i].spines['right'].set_visible(False)

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

## **Pie Chart**

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Assuming 'Branch' is a categorical variable
percentage_distribution_branch = df['Branch'].value_counts(normalize=True) * 100

# Define a custom gradient color scheme
colors = [(0.6, 0.1, 0.1), (0.9, 0.3, 0.3), (0.8, 0.5, 0.5)]

# Create a custom colormap with the defined colors
cmap = LinearSegmentedColormap.from_list('custom', colors, N=len(percentage_distribution_branch))

# Explode the largest segment (optional)
explode = (0.1, 0, 0)

# Create the pie chart with a gradient color scheme and shadow effect
plt.figure(figsize=(8,5))
wedges, texts, autotexts = plt.pie(percentage_distribution_branch, labels=percentage_distribution_branch.index, autopct='%1.1f%%',
                                  startangle=140, explode=explode, colors=cmap(range(len(percentage_distribution_branch))), shadow=True)

# Customize the labels and text properties
for text, autotext in zip(texts, autotexts):
    text.set(size=12, fontweight='bold', color='black')
    autotext.set(size=12, fontweight='bold', color='white')

# Add a legend with border
legend = plt.legend(percentage_distribution_branch.index, loc="center left", bbox_to_anchor=(1, 0.5))
frame = legend.get_frame()
frame.set_edgecolor('black')

# Add a title with a custom font
plt.title('Percentage Distribution of Branches', fontdict={'fontsize': 16, 'fontweight': 'bold'}, pad=20)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the pie chart
plt.show()
```

## **8. 3D Scatter Plot with 'Customer Type':**

The 3D scatter plot with different colors for 'Customer type' shows how 'Unit Price', 'Quantity', and 'Total' sales vary for different types of customers. This can help identify if there are distinct patterns or preferences among customer types.

```python
# Create a 3D scatter plot with different colors for 'Customer type'
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for customer_type, color in zip(df['Customer type'].unique(), ['b', 'r']):
    subset = df[df['Customer type'] == customer_type]
    ax.scatter(subset['Unit price'], subset['Quantity'], subset['Total'], c=color, label=customer_type)

ax.set_xlabel('Unit Price')
ax.set_ylabel('Quantity')
ax.set_zlabel('Total')
plt.title('3D Scatter Plot with Customer Type')
plt.legend()
plt.show()
```
